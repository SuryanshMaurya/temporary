```python
import pandas as pd

data = {
    "teacher": ["Alice", "Bob", "Charlie", "David"],
    "subject": ["Math", "Science", "History", "English"],
    "experience": [5, 10, 3, 8],
    "salary": [50000, 60000, 45000, 55000],
}

df=pd.DataFrame(data)

# Add a null (None) entry
df.loc[len(df)] = [None, None, None, None]

print(df)
```

```python
new_index=["one", "two", "three", "four"]
df.index=new_index
df
```
